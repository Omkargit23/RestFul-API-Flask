**1. What is a RESTful API? **
  
Ans- A RESTful API is a way for different applications to communicate with each other over the internet using simple rules.

REST stands for Representational State Transfer. It mainly uses HTTP methods like:

GET → to get data
POST → to create data
PUT → to update data
DELETE → to remove data

In a RESTful API, everything is treated as a resource (like users, products, etc.), and each resource has a unique URL.

For example:
If we have a users API:
 -GET /users → get all users
 -GET /users/1 → get user with id 1

So basically, a RESTful API helps apps talk to each other in a simple and standard way using URLs and HTTP methods.


**2.  What is Flask, and why is it popular for building APIs? **

Ans - Flask is a lightweight web framework in Python that is used to build web applications and APIs.

It is called a micro framework because it comes with only the basic features, and you can add extra tools if needed.

Flask is popular for building APIs because:
 -It is simple and easy to learn
 -It requires less code to get started
 -It is very flexible (you can design things your own way)
 -It has good support for creating RESTful APIs

So, developers like Flask because it is fast, simple, and flexible, especially for small to medium API projects.


**3. What are HTTP methods used in RESTful APIs? **

Ans - HTTP methods are used in RESTful APIs to perform different operations on data (resources).

The main HTTP methods are:
GET → used to fetch or read data
POST → used to create new data
PUT → used to update existing data
DELETE → used to remove data

For example:
GET /students → get all students
POST /students → add a new student
PUT /students/1 → update student with id 1
DELETE /students/1 → delete student with id 1

So basically, HTTP methods tell the server what action we want to perform on the data.


**4. What is the purpose of the @app.route() decorator in Flask? **

Ans - The @app.route() decorator in Flask is used to connect a URL with a specific function.

It tells Flask that when a user visits a particular URL, a certain function should run.

For example:

@app.route('/hello')
def hello():
    return "Hello World"

Here, when we open /hello in the browser, the hello() function will be executed and show "Hello World".

So basically, @app.route() is used to define routes (URLs) and link them to functions in a Flask app.


**5.  What is the role of Flask-SQLAlchemy? **

Ans - Flask-SQLAlchemy is an extension of Flask that is used to work with databases easily.

It helps us connect our Flask app to a database and perform operations like:
 -adding data
 -retrieving data
 -updating data
 -deleting data

Instead of writing raw SQL queries, we can use Python classes and objects (ORM – Object Relational Mapping).

For example, we can create a table like this:

class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(100))

So basically, Flask-SQLAlchemy makes database handling simpler and cleaner in Flask applications.


**6.  How do you create a basic Flask application?**

Ans - To create a basic Flask application, we just need a few simple steps.

First, install Flask:

pip install flask

Then create a Python file (for example app.py) and write this code:

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Hello, this is my first Flask app!"

if __name__ == '__main__':
    app.run(debug=True)

Now run the file:

python app.py

After running, open your browser and go to:

http://127.0.0.1:5000/

You will see the message on the screen.

So basically, a basic Flask app includes:

creating a Flask object
defining routes using @app.route()
running the app


**7. How do you return JSON responses in Flask? **

Ans - In Flask, we can return JSON responses using the jsonify() function.

It converts Python data (like dictionary or list) into JSON format automatically.

Example:

from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/data')
def get_data():
    return jsonify({
        "name": "Rahul",
        "age": 22
    })

When we open /data, it will return:

{
  "name": "Rahul",
  "age": 22
}

So basically, jsonify() is used to send data in JSON format from a Flask API.
